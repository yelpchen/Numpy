## 1. 导入 NumPy

```python
import numpy as np

print("NumPy version:", np.__version__)
```

## 2. 准备成绩数据

这里用二维数组表示成绩：

- 每一行代表一个学生
- 每一列代表一个科目

```python
students = np.array(["小明", "小红", "小刚", "小丽", "小强", "小雅"])
subjects = np.array(["语文", "数学", "英语", "科学"])

scores = np.array([
    [88, 92, 85, 90],
    [95, 89, 94, 91],
    [76, 81, 79, 85],
    [90, 96, 93, 97],
    [68, 75, 72, 70],
    [84, 88, 90, 86],
])

print("学生：", students)
print("科目：", subjects)
print("成绩表：")
print(scores)
```

## 3. 查看 ndarray 属性

学习 NumPy 时，先养成看数组属性的习惯：维度、形状、元素数量和数据类型。

```python
print("维度 ndim:", scores.ndim)
print("形状 shape:", scores.shape)
print("元素数量 size:", scores.size)
print("数据类型 dtype:", scores.dtype)
```

## 4. 索引和切片

二维数组索引格式：`数组[行, 列]`。

```python
print("第 1 个学生的所有成绩：", scores[0])
print("第 1 个学生的数学成绩：", scores[0, 1])
print("所有学生的数学成绩：", scores[:, 1])
print("前 3 个学生的语文和数学成绩：")
print(scores[:3, :2])
```

## 5. 按科目统计

`axis=0` 表示沿着行方向压缩，也就是对每一列做统计。这里可以理解为“每个科目的统计”。

```python
subject_mean = scores.mean(axis=0)
subject_max = scores.max(axis=0)
subject_min = scores.min(axis=0)

for subject, mean_score, max_score, min_score in zip(subjects, subject_mean, subject_max, subject_min):
    print(f"{subject}: 平均分={mean_score:.1f}, 最高分={max_score}, 最低分={min_score}")
```

## 6. 按学生统计

`axis=1` 表示沿着列方向压缩，也就是对每一行做统计。这里可以理解为“每个学生的统计”。

```python
student_total = scores.sum(axis=1)
student_mean = scores.mean(axis=1)

for name, total, mean_score in zip(students, student_total, student_mean):
    print(f"{name}: 总分={total}, 平均分={mean_score:.1f}")
```

## 7. 找出最高分学生和最强科目

```python
best_student_index = np.argmax(student_mean)
best_subject_index = np.argmax(subject_mean)

print("平均分最高的学生：", students[best_student_index])
print("该学生平均分：", student_mean[best_student_index])
print("平均分最高的科目：", subjects[best_subject_index])
print("该科目平均分：", subject_mean[best_subject_index])
```

## 8. 布尔筛选

布尔筛选是 NumPy 很常用的能力。先得到一组 `True` / `False`，再用它筛选数组。

```python
excellent_mask = student_mean >= 90

print("是否优秀：", excellent_mask)
print("优秀学生：", students[excellent_mask])
print("优秀学生平均分：", student_mean[excellent_mask])
```

```python
weak_mask = scores < 80
weak_positions = np.where(weak_mask)

for row, col in zip(weak_positions[0], weak_positions[1]):
    print(f"{students[row]} 的 {subjects[col]} 需要加强：{scores[row, col]} 分")
```

## 9. 排名

`np.argsort` 返回的是排序后的索引。默认从小到大，想从高到低可以加 `[::-1]`。

```python
ranking_index = np.argsort(student_mean)[::-1]

print("平均分排名：")
for rank, index in enumerate(ranking_index, start=1):
    print(f"第 {rank} 名：{students[index]}，平均分 {student_mean[index]:.1f}")
```

## 10. 广播机制

广播可以让不同形状的数组进行批量运算。下面给每个科目设置不同加分。

```python
bonus = np.array([2, 0, 3, 1])
new_scores = scores + bonus

print("原始成绩：")
print(scores)
print("加分后成绩：")
print(new_scores)
```

## 11. 随机模拟练习

用随机数模拟 10 个学生、4 个科目的成绩，再做一次统计。

```python
rng = np.random.default_rng(seed=42)
random_scores = rng.integers(60, 101, size=(10, 4))

print(random_scores)
print("每个学生平均分：", random_scores.mean(axis=1))
print("每个科目平均分：", random_scores.mean(axis=0))
```
