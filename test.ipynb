# Jupyter 常见快捷键

Jupyter Notebook 有两种模式：

- 命令模式：按 `Esc` 进入，用来操作单元格
- 编辑模式：按 `Enter` 进入，用来编辑单元格内容

## 常用快捷键

| 快捷键 | 作用 |
| --- | --- |
| `Esc` | 进入命令模式 |
| `Enter` | 进入编辑模式 |
| `Shift + Enter` | 运行当前单元格，并跳到下一个单元格 |
| `Ctrl + Enter` | 运行当前单元格，停留在当前单元格 |
| `Alt + Enter` | 运行当前单元格，并在下方新建单元格 |
| `A` | 在上方新建单元格 |
| `B` | 在下方新建单元格 |
| `D, D` | 删除当前单元格，连续按两次 `D` |
| `Z` | 撤销删除单元格 |
| `M` | 把当前单元格改成 Markdown |
| `Y` | 把当前单元格改成代码 |
| `X` | 剪切当前单元格 |
| `C` | 复制当前单元格 |
| `V` | 粘贴单元格到下方 |
| `Shift + V` | 粘贴单元格到上方 |
| `L` | 显示或隐藏行号 |
| `H` | 打开快捷键帮助 |
| `0, 0` | 重启内核，连续按两次 `0` |

## Markdown 常用写法

```markdown
# 一级标题
## 二级标题
- 无序列表
1. 有序列表
`行内代码`
```
